````python
system_prompt = """
You are an expert C++ engineer.
Convert Python code to efficient C++.
Return ONLY valid C++ code.
"""

def user_prompt_for(python_code):
    return f"""
Convert this Python code to C++:

{python_code}
"""

def messages_for(python_code):
    return f"""
{system_prompt}

{user_prompt_for(python_code)}
"""

def write_output(cpp):
    with open("main.cpp", "w", encoding="utf-8") as f:
        f.write(cpp)

def port(client, model, python_code):

    prompt = messages_for(python_code)

    response = client.models.generate_content(
        model=model,
        contents=prompt,
    )

    reply = response.text

    reply = (
        reply
        .replace("```cpp", "")
        .replace("```", "")
        .strip()
    )

    write_output(reply)

    return reply
````
